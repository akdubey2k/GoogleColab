```python
import pdb
#pdb.set_trace()

# Python program for implementation of MergeSort
def mergeSort(arr):
  
	if len(arr) > 1:
       # arr = [12, 11, 13, 5, 6, 7, 23]

		# Finding the mid of the array
		mid = len(arr)//2
  # mid = 3

		# Dividing the array elements
		L = arr[:mid]
  # L = [12, 11, 13]

		# into 2 halves
		R = arr[mid:]
  # R = [5, 6, 7, 23]

		# Sorting the first half
		mergeSort(L)

		# Sorting the second half
		mergeSort(R)
  # L = [11, 12, 13], R = [5, 6, 7, 23]

		i = j = k = 0

		# Copy data to temp arrays L[] and R[]
		while i < len(L) and j < len(R):
			if L[i] < R[j]:
				arr[k] = L[i]
				i += 1
			else:
				arr[k] = R[j]
				j += 1
			k += 1

		# Checking if any element was left
		while i < len(L):
			arr[k] = L[i]
			i += 1
			k += 1

		while j < len(R):
			arr[k] = R[j]
			j += 1
			k += 1

# Code to print the list
def printList(arr):
	for i in range(len(arr)):
		print(arr[i], end=" ")
	print()


# Driver Code
if __name__ == '__main__':
	arr = [12, 11, 13, 5, 6, 7, 23]
	print("Given array is", end="\n")
	printList(arr)
	mergeSort(arr)
	print("Sorted array is: ", end="\n")
	printList(arr)
```
