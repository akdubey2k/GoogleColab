# Quick Sort

Like Merge Sort, QuickSort is a Divide and Conquer algorithm. It picks an element as pivot and partitions the given array around the picked pivot. There are many different versions of quickSort that pick pivot in different ways. 

1. Always pick first element as pivot.
2. Always pick last element as pivot.
3. Pick a random element as pivot.
4. Pick median as pivot.

# 1. Always pick first element as pivot.

```python
import pdb
#pdb.set_trace()

# Python3 implementation of QuickSort 
 
# This Function handles sorting part of quick sort, start and end points to 
# first and last element of an array respectively
def partition(start, end, array):
                        # array = [ 10, 7, 8, 9, 1, 5 ]
    # Initializing pivot's index to start
    pivot_index = start
                # start = 0
    # picking pivot as first element i.e. starting from  left
    pivot = array[pivot_index]
  # pivot = 10
     
    # This loop runs till start pointer crosses end pointer, and when it does we
    # swap the pivot with element on end pointer
    while start < end:
         
      # Increment the start pointer till it finds an element greater than  pivot
        while start < len(array) and array[start] <= pivot:
            start += 1 
          # start = 1, 2, 3, 4, 5
        # array[start] = 7, 8, 9, 1, 5
        # pivot = 10
        # len(array) = 6
             
        # Decrement the end pointer till it finds an element less than pivot
        while array[end] > pivot:
            end -= 1
        #print("end".ljust(25, " "),end) 
        # If start and end have not crossed each other, swap the numbers on start and end
        if(start < end):
            array[start], array[end] = array[end], array[start]
     
    # Swap pivot element with element on end pointer.
    # This puts pivot on its correct sorted place.
    array[end], array[pivot_index] = array[pivot_index], array[end]
                                # array[pivot_index]= 10, array[end]= 5
    # array = [5, 7, 8, 9, 1, 10]
    # Returning end pointer to divide the array into 2
    return end
     
# The main function that implements QuickSort
def quick_sort(start, end, array):
     
    if (start < end):
        # p is partitioning index, array[p] is at right place
        p = partition(start, end, array)
      # p = 5 
         
        # Sort elements before partition and after partition
        quick_sort(start, p - 1, array)
                # (   0 ,  4   , [5, 7, 8, 9, 1, 10])
        quick_sort(p + 1, end, array)
         
# Driver code
# array = [ 10, 7, 8, 9, 1, 5 ]
array = [-10, -7, -8, -9, -1, -5]
# start = 0, end = 5
print("start".ljust(25, " "),0)
print("end".ljust(25, " "),len(array)-1)

quick_sort(0, len(array) - 1, array)
 
print(f'Sorted array: {array}')
```

# 2. Always pick last element as pivot.

```python
# Python program for implementation of Quicksort Sort
 
# This function takes last element as pivot, places the pivot element at its 
# correct position in sorted array, and places all smaller (smaller than pivot) 
# to left of pivot and all greater elements to right of pivot
 
def partition(arr, low, high):

    i = (low-1)           # index of smaller element
    pivot = arr[high]     # choose the rightmost element as pivot
 
    # traverse through all elements, compare each element with pivot
    for j in range(low, high):
 
        # If current element is smaller than or equal to pivot
        # if the element smaller than pivot is found swap it with the greater element pointed by i
        if arr[j] <= pivot:
 
            # increment index of smaller element
            i = i+1
            # swapping element at i with element at j
            arr[i], arr[j] = arr[j], arr[i]

    # swap the pivot element with the greater element specified by i
    arr[i+1], arr[high] = arr[high], arr[i+1]
    return (i+1)
 
# The main function that implements QuickSort
# arr[] --> Array to be sorted,
# low  --> Starting index,
# high  --> Ending index
 
# Function to do Quick sort
def quickSort(arr, low, high):
    if len(arr) == 1:
        return arr
    if low < high:
 
        # pi is partitioning index, arr[p] is now at right place
        pi = partition(arr, low, high)
 
        # Separately sort elements before partition and after partition
        quickSort(arr, low, pi-1)
        quickSort(arr, pi+1, high)
 
 
# Driver code to test above
# arr = [10, 7, 8, 9, 1, 5]
arr = [-10, -7, -8, -9, -1, -5]
n = len(arr)
quickSort(arr, 0, n-1)
print("Sorted array is:", arr)

# for i in range(n):
#     print("%d" % arr[i])
```
