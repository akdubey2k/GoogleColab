# Iterative Binary Search

```python
# It returns index position of an "element" in a given list "arr" if present, else returns -1   
def binary_search(arr, element):
  low = mid = 0
  high = len(arr) - 1  

  while (low <= high):
    # for get integer result
    # mid = int(low + (high-low)/2) # // 2
    mid = (low + high) // 2  

    # If element is greater, compare to the right of mid
    if arr[mid] < element:  
      low = mid + 1  

    # If element is smaller, compared to the left of mid
    elif arr[mid] > element:  
      high = mid - 1  

    # Check if element is present at mid  
    else:  
      return mid  

    # element was not present in the list, return -1  
  return -1  
  
  
# Initial list array
arr = [12, 24, 32, 39, 45, 50, 54]
# arr = [12,12,12,12,12,12,12,12,12,12]
element = 32  
  
# Function call   
result = binary_search(arr, element)  
  
if result != -1:  
    print("Element is present at index", str(result))  
else:  
    print("Element is not present in list")  
```

# Recursive Binary Search

```python
# Returns index position of an "element" in a list "arr" if present, otherwise -1  
def binary_search(arr, low, high, element):   
  
   # Check base case for the recursive function  
   if low <= high:  
  
      mid = (low + high) // 2  
  
      # If element is available at the middle itself then return the its index  
      if arr[mid] == element:   
         return mid   
  
      # If the element is smaller than mid value, then search moves left sublist  
      elif arr[mid] > element:   
         return binary_search(arr, low, mid-1, element)   
  
      # Else the search moves to the right sublist  
      else:   
         return binary_search(arr, mid+1, high, element)   
  
   else:   
      # Element is not available in the list  
      return -1  
  
# Test list1   
#arr = [12, 24, 32, 39, 45, 50, 54]  
arr = [12,12,12,12,12,12,12,12,12,12]
element = 12  
  
# Function call   
res = binary_search(arr, 0, len(arr)-1, element)   
  
if res != -1:   
   print("Element is present at index", str(res))  
else:   
   print("Element is not present in list")  
```

```python
from typing import List # For annotations

def BinarySearch (arr : List[int], element : int) :

    beg = 0
    end = len(arr) - 1 

    while (beg <= end) :

        mid = int(beg + (end-beg)/2)

        if (arr[mid] == element) :
            return mid 
        elif (arr[mid] < element) :
            beg = mid + 1;  
        else :
            end = mid - 1

    return -1; 

arr = [ 3, 5, 7, 11, 12, 16, 17, 52, 64, 101 ]

print("Array : " + str(arr))
# wait for input element to be search in binary search algorithm
element = int(input("Element to search : "))
index = BinarySearch (arr, element)

if (index != -1) :
    print("Element found at index : " + str(index))
else :
    print("Element not found. ")
```
